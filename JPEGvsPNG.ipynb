```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
data_dir = '/mnt/ssd1/biomarker/c-met/final_output/check'
data_list = sorted(os.listdir(data_dir))

png_list = [i for i in data_list if 'input.png' in i]
jpeg_75_list = [i for i in data_list if 'input.jpg' in i]
jpeg_80_list = [i for i in data_list if 'input_80.jpg' in i]
jpeg_85_list = [i for i in data_list if 'input_85.jpg' in i]
jpeg_90_list = [i for i in data_list if 'input_90.jpg' in i]
```

```python
def get_file_size_kB(data_dir, file):
    return round(os.path.getsize(os.path.join(data_dir, file)) / 1000, 1)

file_ids = []
file_sizes = []

for i in range(len(png_list)):

    if not (png_list[i].split('input')[0] == jpeg_75_list[i].split('input')[0] and png_list[i].split('input')[0] == jpeg_80_list[i].split('input')[0] \
        and png_list[i].split('input')[0] == jpeg_85_list[i].split('input')[0] and png_list[i].split('input')[0] == jpeg_90_list[i].split('input')[0]):
        print('Different files')
        break

    file_ids.append(png_list[i].split('_input')[0])
    file_sizes.append([get_file_size_kB(data_dir, png_list[i]), get_file_size_kB(data_dir, jpeg_75_list[i]), get_file_size_kB(data_dir, jpeg_80_list[i]), get_file_size_kB(data_dir, jpeg_85_list[i]), get_file_size_kB(data_dir, jpeg_90_list[i])])
```

```python
file_ids
```

```python
file_sizes = np.array(file_sizes)
file_sizes
```
